### Feature Selection by using three main methods

Now, we will decide which features have good correlations with target feature "Revenue", we will use three ways to do this: 
- Univariate selection
- Tree-based feature selection
- Using a correlation matrix

These three ways are main methods to check whether the discriptive features are good features to target

```python
trainset = dataset.drop('Revenue',axis=1)
target = dataset['Revenue']
```

#### Univariate Selection

Univariate feature selection works by selecting the best feature based on univariate statistical test

```python
best_features = SelectKBest(score_func=chi2, k='all').fit(trainset, target)

dfscores = pd.DataFrame(best_features.scores_)
dfcolumns = pd.DataFrame(trainset.columns)

featureScores = pd.concat([dfcolumns,dfscores], axis=1)
featureScores.columns = ['Feature','Score']

print(featureScores.sort_values(by='Score', ascending=False))
```

From this method, we could easily get information that, ProductRelated_Duration, PageValues and Administrative_Duration are the first three good features of this dataset. However, for Browser, Weekend, Region, TrafficType and OperatingSystems, their scores are relative low, they are under 10, in other words, they are not good features.

- By our analysis in last part, we get information that Weekend and Region are not good features. 

- OperatingSystems is a feature that we need to recheck we mentioned in the last part, now, we know that at least in this method, OperatingSystems is not good. 

- In the previous analysis, Browser is a feature whose relationship is not so close with target feature. Now, we are sure of this by this table.

### Correlation matrix
Correlation states how the features are related to each other or the target variable. It can be either positive or 
negative. So the values that are closer to 1 and -1 are really good and values closer to 0 from both sides are not good. 

Here, we use heatmap to identify the correlations between features and target and different feature.

```python
combine = pd.concat([trainset, target], axis=1)
corr_mat = combine.corr()

plt.figure(figsize=(20,20))
g=sns.heatmap(corr_mat,annot=True,cmap="RdYlGn")
```

### K-Nearest Neighbours

KNN is that each sample can be represented by its nearest k neighbor values, which is a method to classify each record in the data set.

#### Curse of Dimensionality
Curse of dimensionality means that the higher the spatial dimension of your dataset(which means that the dataset have many discriptive features), some aspects of learning become more difficult.

For this reason, KNN algorithm cannot have many discriptive features, since it is influenced by dimensionality easily. Dimension will lead that it is hard to judge the neighbours even let you cannot decide what category this feature should be in. 

We will choose features before we train it.

#### Handle Categorical Data
In KNN algorithm, for categorical data, we could not calculate the distance between different values, for this, I need to convert categorical data into binary values.

For Oversampling dataset

```python
ov_ct = pd.concat([oversampling,pd.get_dummies(oversampling['Month'], prefix='Month')], axis=1).drop(['Month'],axis=1)
ov_ct = pd.concat([ov_ct,pd.get_dummies(ov_ct['VisitorType'], prefix='VisitorType')], axis=1).drop(['VisitorType'],axis=1)
```

For Undersampling dataset

```python
un_ct = pd.concat([undersampling,pd.get_dummies(undersampling['Month'], prefix='Month')], axis=1).drop(['Month'],axis=1)
un_ct = pd.concat([un_ct,pd.get_dummies(un_ct['VisitorType'], prefix='VisitorType')], axis=1).drop(['VisitorType'],axis=1)
```

#### Hyperparameter we need to consider

In KNN, there are three hyperparameters we need to consider: 
1. n_neighbors: Number of neighbors that we decide to use in the train 
2. p: Determins the method of calculating distance
    - p = 1 ==> manhattan_distance
    - p = 2 ==> euclidean_distance
3. weights: Weight of features used in prediction

#### For Oversampling dataset

Split target and discriptive features

```python
X = ov_ct.drop('Revenue',axis=1)
y = ov_ct['Revenue']
ov_ct.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.4)
X_train_cat = X_train[['TotalDuration','ExitRates', 'PageValues','Month_2', 'Month_3','Month_5','Month_6',
                       'Month_7','Month_8','Month_9','Month_10','Month_11','Month_12','VisitorType_0',
                      'VisitorType_1','VisitorType_2']]

# 'Month_2', 'Month_3','Month_5','Month_6',
#                        'Month_7','Month_8','Month_9','Month_10','Month_11','Month_12'
X_valid_cat = X_test[['TotalDuration','ExitRates', 'PageValues','Month_2', 'Month_3','Month_5','Month_6',
                       'Month_7','Month_8','Month_9','Month_10','Month_11','Month_12','VisitorType_0',
                      'VisitorType_1','VisitorType_2']]

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
pred = knn.predict(X_test)
default_acc = metrics.accuracy_score(y_test, pred)
print("Accuracy for Intial KNN accuracy:", default_acc)
print("*"*40)
print(classification_report(y_pred=pred, y_true=y_test))
```

### Hyperparameter Tuning  for Oversampling Dataset

We will use different p values to check accuracies by increasing the number of the nearest neighbors and we also use different modes of weights. So there are four lines in this graph.

```python
p1_accuracy_un = []
p2_accuracy_un = []
p1_accuracy_dis = []
p2_accuracy_dis = []
for n in range(1,30):
    X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.4)
    X_train_cat = X_train[['TotalDuration','ExitRates', 'PageValues','Month_2', 'Month_3','Month_5','Month_6',
                           'Month_7','Month_8','Month_9','Month_10','Month_11','Month_12','VisitorType_0',
                          'VisitorType_1','VisitorType_2']]

    # 'Month_2', 'Month_3','Month_5','Month_6',
    #                        'Month_7','Month_8','Month_9','Month_10','Month_11','Month_12'
    X_valid_cat = X_test[['TotalDuration','ExitRates', 'PageValues','Month_2', 'Month_3','Month_5','Month_6',
                           'Month_7','Month_8','Month_9','Month_10','Month_11','Month_12','VisitorType_0',
                          'VisitorType_1','VisitorType_2']]

    p1_knn_un = KNeighborsClassifier(n_neighbors=n, p=1,weights = 'uniform')
    p1_knn_un.fit(X_train, y_train)
    p1_pred_un = p1_knn_un.predict(X_test)
    p1_accuracy_un.append(metrics.accuracy_score(y_test, p1_pred_un))
    
    p1_knn_dis = KNeighborsClassifier(n_neighbors=n, p=1,weights = 'distance')
    p1_knn_dis.fit(X_train, y_train)
    p1_pred_dis = p1_knn_dis.predict(X_test)
    p1_accuracy_dis.append(metrics.accuracy_score(y_test, p1_pred_dis))
    
    p2_knn_un = KNeighborsClassifier(n_neighbors=n, p=2, weights = 'uniform')
    p2_knn_un.fit(X_train, y_train)
    p2_pred_un = p2_knn_un.predict(X_test)
    p2_accuracy_un.append(metrics.accuracy_score(y_test, p2_pred_un))
    
    p2_knn_dis = KNeighborsClassifier(n_neighbors=n, p=2, weights = 'distance')
    p2_knn_dis.fit(X_train, y_train)
    p2_pred_dis = p2_knn_dis.predict(X_test)
    p2_accuracy_dis.append(metrics.accuracy_score(y_test, p2_pred_dis))
plt.figure(figsize=(20, 6))
plt.plot(range(1, 30), p1_accuracy_un, color='red', marker='o',
         markersize=10,label='p=1,uniform')
plt.plot(range(1, 30), p2_accuracy_un, color='blue', marker='o',
         markersize=10,label='p=2,uniform')
plt.plot(range(1, 30), p1_accuracy_dis, color='orange', marker='o',
         markersize=10,label='p=1,distance')
plt.plot(range(1, 30), p2_accuracy_dis, color='purple', marker='o',
         markersize=10,label='p=2,distance')
plt.legend()
plt.title('Accuracy Vs K Value')
plt.xlabel('K Value')
plt.ylabel('Accuracy')
```

### Linear regression

It is a generalized linear model, so it has many similarities with multiple linear regression analysis. Their model forms are basically the same with Ax + B, where A and B are parameters to be solved, and the difference lies in their dependent variables

#### Hyperparameter we need to consider

1. C: Inverse of regularization strength
2. solver:Algorithm to use in the optimization problem.
    - For small datasets, liblinear is a good choice, whereas sag and saga are faster for large ones.
    - For multiclass problems, only newton-cg, sag, saga and lbfgs handle multinomial loss; liblinear is limited to one-versus-rest schemes.
3. max_iter: Maximum number of iterations taken for the solvers to converge.

## Ensemble

In the supervised learning algorithm of machine learning, our goal is to learn a stable model with good performance in all aspects, but the actual situation is often not so ideal. Sometimes we can only get multiple models with preference (weak supervised model, which performs better in some aspects). Ensemble learning is to combine several weak supervised models to get a better and more comprehensive strong supervised model. The potential idea of ensemble learning is that even if one weak classifier gets a wrong prediction, other weak classifiers can correct the error.

In this section, I decide to use the model with high accuracy and its responded dataset.

### Bagging

The core though of bagging is given a weak learning algorithm and a training set, the learning algorithm is used many times to get the prediction function sequence and vote. Finally, the accuracy of the result will be improved

#### Hyperparameter we need to consider
1. base_estimator: 
    - The base estimator to fit on random subsets of the dataset. If None, then the base estimator is a decision tree.
2. n_estimators: 
    - The number of base estimators in the ensemble.

Set the dataset for each training model

```python
X = oversampling.drop('Revenue',axis=1)
y = oversampling['Revenue']

X_KNN = ov_ct.drop('Revenue',axis=1)
y_KNN = ov_ct['Revenue']

X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.4)

X_train_k, X_test_k, y_train_k, y_test_k = train_test_split(X_KNN, y_KNN , test_size=0.4)

X_train_cat = X_train[['TotalDuration','ExitRatesCate', 'PageValues', 'Month', 'TrafficType','VisitorType']]
X_valid_cat = X_test[['TotalDuration','ExitRatesCate', 'PageValues', 'Month', 'TrafficType','VisitorType']]

X_train_KNN = X_train_k[['TotalDuration','ExitRates', 'PageValues','Month_2', 'Month_3','Month_5','Month_6',
                       'Month_7','Month_8','Month_9','Month_10','Month_11','Month_12','VisitorType_0',
                      'VisitorType_1','VisitorType_2']]
X_valid_KNN = X_test_k[['TotalDuration','ExitRates', 'PageValues','Month_2', 'Month_3','Month_5','Month_6',
                       'Month_7','Month_8','Month_9','Month_10','Month_11','Month_12','VisitorType_0',
                      'VisitorType_1','VisitorType_2']]
```

### Hyperparameter Tuning for Bagging

```python
log_acc=[]
des_acc=[]
knn_acc=[]
gnb_acc=[]
for n in range(1,20):
    base_log_reg = LogisticRegression(C=100, solver = 'newton-cg', max_iter=10)
    bag_clf = BaggingClassifier(base_estimator=base_log_reg, n_estimators=n)
    bag_clf.fit(X_train_cat, y_train)
    pred = bag_clf.predict(X_valid_cat)
    log_acc.append(metrics.accuracy_score(y_test, pred))



    base_decision = DecisionTreeClassifier(criterion='gini', max_depth = 30, random_state=42)
    bag_clf = BaggingClassifier(base_estimator=base_decision, n_estimators=n)
    bag_clf.fit(X_train_cat, y_train)
    pred = bag_clf.predict(X_valid_cat)
    des_acc.append(metrics.accuracy_score(y_test, pred))



    base_KNN = KNeighborsClassifier(p=1,weights='distance',n_neighbors=1)
    bag_clf = BaggingClassifier(base_estimator=base_KNN, n_estimators=n)
    bag_clf.fit(X_train_cat, y_train)
    pred = bag_clf.predict(X_valid_cat)
    knn_acc.append(metrics.accuracy_score(y_test, pred))

    base_GNB = GaussianNB(var_smoothing = 1e-20)
    bag_clf = BaggingClassifier(base_estimator=base_GNB, n_estimators=n)
    bag_clf.fit(X_train_cat, y_train)
    pred = bag_clf.predict(X_valid_cat)
    gnb_acc.append(metrics.accuracy_score(y_test, pred))

plt.rcParams['figure.figsize'] = (20, 10)
plt.plot(range(1,20), log_acc, color='blue', marker='o',
         markersize=10,label = 'logistic')
plt.plot(range(1,20), des_acc, color='red', marker='o',
         markersize=10,label = 'Decision Tree')
plt.plot(range(1,20), knn_acc, color='purple', marker='o',
         markersize=10,label = 'KNN')
plt.plot(range(1,20), gnb_acc, color='grey', marker='o',
         markersize=10,label = 'Naive Bayes')
plt.legend()
plt.title('Accuracy Vs n_estimators')
plt.xlabel('n_estimatorsvalue')
plt.ylabel('Accuracy')
```

### Voting Classifier

Voting is a voting mechanism, which is divided into soft voting and hard voting. Its principle adopts the idea that the minority is subordinate to the majority.

#### Hyperparameter we need to consider
1. weights: 
    - Sequence of weights (float or int) to weight the occurrences of predicted class labels (hard voting) or class probabilities before averaging (soft voting). Uses uniform weights if None.
2. voting(hard,soft):
    - If hard, uses predicted class labels for majority rule voting. Else if soft, predicts the class label based on the argmax of the sums of the predicted probabilities, which is recommended for an ensemble of well-calibrated classifiers.

Here, we just use three high accuracy algorithm which are Decision Tree, Logistic Regreesion and K-Neares Neighbors to figure out the best parameters for Voting Classifier.

Firstly, Let's check the accuracy without doing hyperparameter tuning

```python
tree_clf = DecisionTreeClassifier(criterion='gini', max_depth = 30, random_state=42)
log_reg_clf = LogisticRegression(C=100, solver = 'newton-cg', max_iter=10)
knn_clf = KNeighborsClassifier(p=1,weights='distance',n_neighbors=1)


voting_clf = VotingClassifier([('decTree', tree_clf), ('LogReg', log_reg_clf), ('KNN', knn_clf)])
voting_clf.fit(X_train, y_train)
pred = voting_clf.predict(X_test)
print("Accuracy before tuning Hyperparameter:",metrics.accuracy_score(y_test, pred)*100,"%")
print("*"*40)
print(classification_report(y_pred=pred, y_true=y_test))
```

### Hyperparameter Tuning for Voting Classifier

Verify the influence of weights on the accuracy of the algorithm

```python
soft_acc = []
hard_acc = []
vt_cl = {}
for i in range(1,4):
    for j in range(1,4):
        for m in range(1,4):
            voting_clf = VotingClassifier([('decTree', tree_clf), ('LogReg', log_reg_clf), ('KNN', knn_clf)],weights=[i,j,m])
            voting_clf.fit(X_train, y_train)
            pred = voting_clf.predict(X_test)
            print("Weights is :",i , j, m,". The accuracy is:", metrics.accuracy_score(y_test, pred))
            vt_cl[str(i)+","+str(j)+","+str(m)] = metrics.accuracy_score(y_test, pred)
```

```python
sorted(vt_cl.items(),key=lambda item:item[1],reverse=True)[0:5]
```

From this we could see that the first five accuracy is the same, and the weights are different. As far as I consider, the weights are the voting weight of different estimators(algorithms), so here, by our analysis before, we know that Decision Tree is the best algorithm. So we choose (3,1,2), we let the Decision Tree occupies the most weight.

Now, let's figure out the influence of voting on the accuracy. We will use weights = [3,1,2]

```python
X = oversampling.drop('Revenue',axis=1)
y = oversampling['Revenue']
hard_acc = []
soft_acc = []
for i in range(0,30):  
    X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.4)

    voting_clf_hard = VotingClassifier([('decTree', tree_clf), ('LogReg', log_reg_clf), ('KNN', knn_clf)],weights=[3,1,2],voting='hard')
    voting_clf_hard.fit(X_train, y_train)
    hard_pred = voting_clf_hard.predict(X_test)
    hard_acc.append(metrics.accuracy_score(y_test, hard_pred))
    
    voting_clf_soft = VotingClassifier([('decTree', tree_clf), ('LogReg', log_reg_clf), ('KNN', knn_clf)],weights=[3,1,2],voting='soft')
    voting_clf_soft.fit(X_train, y_train)
    soft_pred = voting_clf_soft.predict(X_test)
    soft_acc.append(metrics.accuracy_score(y_test, soft_pred))
plt.plot(range(0,30), hard_acc, color='purple', marker='o',
         markersize=10,label = 'Hard Voting')
plt.plot(range(0,30), soft_acc, color='red', marker='o',
         markersize=10, label = 'Soft Voting')
plt.legend()
plt.title('Accuracy Vs Voting mode')
plt.xlabel('Test Time')
plt.ylabel('Accuracy')
```

This graph tells us that, soft Voting mode's performance is worse than the Hard one. So for this dataset, I decide to take Hard mode of voting as this algorithm's parameter. Since hard voting is just calculate the number of results, the final result is the voted one which is the bigger number. So in this dataset, hard voting is better than soft.
